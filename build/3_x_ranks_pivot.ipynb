```python
events = ['333', '444', '555', '222', '333bf', '333oh', '333fm', '333ft', 'minx',
              'pyram', 'sq1', 'clock', 'skewb', '666', '777', '444bf', '555bf', '333mbf']
```

```python
for eventId in events:
    print "-- %s" % eventId
    print "best%s INT DEFAULT 0 NOT NULL," % (eventId)
    print "worldRank%s INT DEFAULT 0 NOT NULL," % (eventId)
    print "continentRank%s INT DEFAULT 0 NOT NULL," % (eventId)
    print "countryRank%s INT DEFAULT 0 NOT NULL," % (eventId)
    print "worldCentile%s TINYINT DEFAULT 0 NOT NULL," % (eventId)
    print "continentCentile%s TINYINT DEFAULT 0 NOT NULL," % (eventId)
    print "countryCentile%s TINYINT DEFAULT 0 NOT NULL," % (eventId)
```

```python
for eventId in events:
    print "-- %s" % eventId
    print "CAST(IFNULL(MIN(IF(eventId = '%s', best, NULL)), 0) AS UNSIGNED INT) AS best%s," % (eventId, eventId)
    print "CAST(IFNULL(MIN(IF(eventId = '%s', worldRank, NULL)), 0) AS UNSIGNED INT) AS worldRank%s," % (eventId, eventId)
    print "CAST(IFNULL(MIN(IF(eventId = '%s', continentRank, NULL)), 0) AS UNSIGNED INT) AS continentRank%s," % (eventId, eventId)
    print "CAST(IFNULL(MIN(IF(eventId = '%s', countryRank, NULL)), 0) AS UNSIGNED INT) AS countryRank%s," % (eventId, eventId)
    print "CAST(IFNULL(MIN(IF(eventId = '%s', worldCentile, NULL)), 0) AS UNSIGNED INT) AS worldCentile%s," % (eventId, eventId)
    print "CAST(IFNULL(MIN(IF(eventId = '%s', continentCentile, NULL)), 0) AS UNSIGNED INT) AS continentCentile%s," % (eventId, eventId)
    print "CAST(IFNULL(MIN(IF(eventId = '%s', countryCentile, NULL)), 0) AS UNSIGNED INT) AS countryCentile%s," % (eventId, eventId)
```
