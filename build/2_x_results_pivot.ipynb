```python
events = ['333', '444', '555', '222', '333bf', '333oh', '333fm', '333ft', 'minx',
              'pyram', 'sq1', 'clock', 'skewb', '666', '777', '444bf', '555bf', '333mbf']
```

```python
for eventId in events:
    print "-- %s" % eventId
    print "numComp%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numFinal%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numPodium%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numRound%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numAttempt%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numSuccess%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numDns%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numDnf%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numSingleWr%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numAverageWr%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numSingleCr%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numAverageCr%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numSingleNr%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "numAverageNr%s SMALLINT DEFAULT 0 NOT NULL," % (eventId)
    print "totTime%s BIGINT DEFAULT 0 NOT NULL," % (eventId)
```

```python
for eventId in events:
    print "-- %s" % eventId
    print "CAST(SUM(IF(eventId = '%s', numComp, 0)) AS UNSIGNED INT) AS numComp%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numFinal, 0)) AS UNSIGNED INT) AS numFinal%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numPodium, 0)) AS UNSIGNED INT) AS numPodium%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numRound, 0)) AS UNSIGNED INT) AS numRound%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numAttempt, 0)) AS UNSIGNED INT) AS numAttempt%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numSuccess, 0)) AS UNSIGNED INT) AS numSuccess%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numDns, 0)) AS UNSIGNED INT) AS numDns%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numDnf, 0)) AS UNSIGNED INT) AS numDnf%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numSingleWr, 0)) AS UNSIGNED INT) AS numSingleWr%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numAverageWr, 0)) AS UNSIGNED INT) AS numAverageWr%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numSingleCr, 0)) AS UNSIGNED INT) AS numSingleCr%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numAverageCr, 0)) AS UNSIGNED INT) AS numAverageCr%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numSingleNr, 0)) AS UNSIGNED INT) AS numSingleNr%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', numAverageNr, 0)) AS UNSIGNED INT) AS numAverageNr%s," % (eventId, eventId)
    print "CAST(SUM(IF(eventId = '%s', totTime, 0)) AS UNSIGNED INT) AS totTime%s," % (eventId, eventId)
```
